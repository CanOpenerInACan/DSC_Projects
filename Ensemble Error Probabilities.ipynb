```python
from scipy.stats import binom
import matplotlib.pyplot as plt
import numpy as np
import math
```

## Ensemble with 11 models and 0.2 error rate

```python
n, p = 11, 0.2
```

```python
print('The probability of 6 models making incorrect prediction: ' + str(binom.pmf(6,n,p, loc=0)))
print('The probability of 7 models making incorrect prediction: ' + str(binom.pmf(7,n,p, loc=0)))
print('The probability of 8 models making incorrect prediction: ' + str(binom.pmf(8,n,p, loc=0)))
print('The probability of 9 models making incorrect prediction: ' + str(binom.pmf(9,n,p, loc=0)))
print('The probability of 10 models making incorrect prediction: ' + str(binom.pmf(10,n,p, loc=0)))
print('The probability of 11 models making incorrect prediction: ' + str(binom.pmf(11,n,p, loc=0)))
print('____________________________________________________________________________________\nThe probability of our ensemble making an incorrect prediction: ' + str(1 - binom.cdf(5,n,p,loc=0)))
```

```python
#Creating CDF and PMF for 11 model, 0.2 percent error
n=11 
p=0.2
lcdf=[]
lpmf=[]
for i in range(1,n+1):
    lcdf.append(binom.cdf(i,n,p))
    lpmf.append(binom.pmf(i,n,p))
```

```python
#Plotting our PMF and CDF
plt.plot(lcdf)
plt.plot(lpmf)
plt.title('Graph of PMF and CDF')
plt.xlabel('Number of Models'), plt.ylabel('Percentage')
plt.show()
```

## Making a function to solve the above probabilities easier

```python
def incorrect(n,p):
    k=math.floor(n/2)
    print('The probability of our ensemble making an incorrect prediction: ' + str('{:.4}'.format(binom.sf(k,n,p))))
```

```python
incorrect(11,0.2)
```

```python
incorrect(11,0.49)
```

```python
incorrect(21,0.49)
```

**Write-Up:** In the end, using an ensemble method greatly reduces our chances of making an incorrect prediction. When we have a bunch of good models that only make mistakes $1/5$ times, including an ensemble of 11 of these reduces that 0.20 down to 0.01165, a significant improvement. When we have a bunch of models that are barely better than a random guess at 0.49 percent error, if we include 11 or 21 of these models, we see how much it helps. This reduces down to 0.4729 with 11 models and 0.463 with 21 models. When the prediction is razor thin like this, being able to create an ensemble to keep reducing that likelihood or making an error is important.
